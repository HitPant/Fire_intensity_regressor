#### Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
#### Weather data observations
#### Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
#### RH : Relative Humidity in %: 21 to 90
#### Ws :Wind speed in km/h: 6 to 29
#### Rain: total day in mm: 0 to 16.8
#### FWI Components
#### Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
#### Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
#### Drought Code (DC) index from the FWI system: 7 to 220.4
#### Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
#### Buildup Index (BUI) index from the FWI system: 1.1 to 68
#### Fire Weather Index (FWI) Index: 0 to 31.1
#### Classes: two classes, namely â€œFireâ€ and â€œnot Fireâ€

```python
# importing libs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset

data = pd.read_csv('dataset/Algerian_forest_fires_dataset_UPDATE.csv', header=1)
data.head()
```

```python
data.info()
```

## Data Cleaning

```python
# find missing(null) values

# data.isnull().sum()
data[data.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
data.loc[:122, "Region"] = 0 # all records before index 122: 0 
data.loc[122:, "Region"] = 1 # all records after index 122: 1
```

```python
data.info()
```

```python
data[['Region']] = data[['Region']].astype(int)
```

```python
data.isnull().sum()
```

```python
# drop null values
data = data.dropna().reset_index(drop=True)
data.head()
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
# removing 122nd row asit has the same column name as row 1
data = data.drop(122).reset_index(drop=True)
data.iloc[[122]]
```

```python
# remove whitespaces in the column name
data.columns = data.columns.str.strip()
data.columns
```

### we need to convert the datatype for certain columns as most all of them are object type

### we will convert the following columns to integer type
month <br>
day <br>
year <br>
Temperature <br>
RH <br>
Ws <br>

```python
data[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = data[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
data.info()
```

### We will convert the columns with object datatype into float except the Classes column as it represents category

```python
# capturing all the object type columns

objects = [features for features in data.columns if data[features].dtypes=='O']

# convert all columns to float except Classes
for i in objects:
    if i != 'Classes':
        data[i]=data[i].astype(float)
        
data.info()
```

```python
data.describe()
```

### save the dataset into a new csv file after cleaing

```python
data.to_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

## Exploratory Data Analysis

```python
# dropping columns and perform EDA using rest of the columns
df = data.drop(['day','year','month'], axis=1)
df.head()
```

```python
# Encode categorical variables: data['Classes']
# where it is not fire convert it to 0 and rest 1

df['Classes']=np.where(df['Classes'].str.contains('not fire'), 0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
# density plot for all features
plt.style.use('seaborn')
df.hist(bins=50, figsize=(20,15))
```

```python
percentage = df['Classes'].value_counts(normalize=True)*100

classlabels = ['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
df.corr()
```

```python
data['Classes']=np.where(data['Classes'].str.contains('not fire'), 'not fire','fire')
```

```python
dftemp = data.loc[data['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=data)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Monthly fire analysis of Sidi-Bel Region')
```

### - fire started to increase from month of July.
### - August had the highest number of fires.

```python
dftemp = data.loc[data['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=data)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Monthly fire analysis of Bejai Region')
```

### Refering to both the region, it can be infered that:

#### In the month of June, July and August the number of fire started to increase and decreased in month of September
